IMPORT Libraries 

```python
import featuretools as ft #import feature tools for running featuring engineering on data 
import pandas as pd #import pandas for data manipulation 
import numpy as np #numerical computation
```

Load the data on ecommerce ( https://www.kaggle.com/datasets/olistbr/brazilian-ecommerce )

Customer dataframe

```python
cx_df=pd.read_csv('/Users/harishs/Documents/Ecom/Data/Master/olist_customers_dataset.csv')
```

Payment dataframe 

```python
pay_df=pd.read_csv('/Users/harishs/Documents/Ecom/Data/Master/olist_order_payments_dataset.csv')
```

Orders dataframe

```python
ord_df=pd.read_csv('/Users/harishs/Documents/Ecom/Data/Master/olist_orders_dataset.csv')
```

Create an entity set which is a feature from feature tools library , basically it defines the relationship between dataframes

```python
es=ft.EntitySet(id='cx_ord_pay')
```

Add the first dataframe to the entity set

```python
es.add_dataframe(dataframe_name='cxdata',dataframe=cx_df,index='customer_id')
```

As order_id is not unique , we are supposed to create a dummy index column 


```python
pay_df["pay_index"] = pay_df.index
es = es.add_dataframe(dataframe_name="paydata", dataframe=pay_df, index="pay_index")
```

Add third dataframe to entity set

```python
es.add_dataframe(dataframe_name='orderdata',dataframe=ord_df,index='order_id')
```

DEFINE PARENT AND CHILD DATAFRAME AND ITS INDICES

```python
cx_order = ft.Relationship(
    entityset=es,
    parent_dataframe_name="cxdata",
    parent_column_name="customer_id",
    child_dataframe_name="orderdata",
    child_column_name="customer_id",
)
```

```python
cx_pay=ft.Relationship(
    entityset=es,
    parent_dataframe_name="orderdata",
    parent_column_name="order_id",
    child_dataframe_name="paydata",
    child_column_name="order_id",
)
```

ADD RELATIONSHIP BETWEEN DATAFRAME

```python
es.add_relationship(relationship=cx_order)
```

```python
es.add_relationship(relationship=cx_pay)
```

RUN Deep Feature Synthesis to produce new columns

```python
# Generate features
feature_matrix, feature_defs = ft.dfs(entityset=es, target_dataframe_name="cxdata")
```
